```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
```

```python
# reading pickle files into pandas dataframes
PoliceBX_tweets = pd.read_pickle("./PoliceCommBX_tweets.pkl")
BX_tweets = pd.read_pickle("./BX_tweets.pkl")
BK_tweets = pd.read_pickle("./BK_tweets.pkl")
SI_tweets = pd.read_pickle("./SI_tweets.pkl")
BK_tweets = BK_tweets[BK_tweets.nlikes>=1]
```

```python
from nltk.corpus import stopwords

PoliceBX_stopwords = ['www','https','https twitter','https twitter com','twitter',
                'pic twitter','pic twitter com','twitter com',
                'https www', 'https www instagram','instagram','instagram com',
                'www instagram','www instagram com','at the','in the','of the',
                'the protest','if you','this is','on the','to the','you re']

# countvectorizing PoliceBX tweets
vectorizer = CountVectorizer(ngram_range=(1,3))

X = vectorizer.fit_transform(PoliceBX_tweets['tweet'])

user_ids = list(PoliceBX_tweets['user_id'])

countvec_PoliceBX = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())

# build df with PoliceBX user_ids and countvectorized tweets
countvec_PoliceBX.insert(0, "user_id", user_ids, False) 

# import list of general stopwords
stopwords = stopwords.words()

# create list of stopwords found in the PoliceBX tweets
stopwords = list(set(countvec_PoliceBX.columns) & set(stopwords))+PoliceBX_stopwords

# remove stopwords from countvectorized PoliceBX tweets
countvec_PoliceBX = countvec_PoliceBX.drop(stopwords,axis=1)
```
